```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
```

```python
# Veri dosyasını oku
df = pd.read_csv('normalized_data.csv')
```

```python
# Eksik değerleri kontrol et ve temizle
print("Eksik değerler:")
print(df.isnull().sum())
df = df.dropna()
```

```python
# Özellikler (X) ve hedef değişken (y) olarak ayır
X = df.drop('Class variable (0 or 1)', axis=1)
y = df['Class variable (0 or 1)']
```

```python
# Veri setini eğitim ve test setlerine ayır (%70 eğitim, %30 test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# MLP sınıflandırıcıyı eğit
mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)
```

```python
# SVM sınıflandırıcıyı eğit
svm = SVC(probability=True, random_state=42)
svm.fit(X_train, y_train)
```

```python
import seaborn as sns
# Performans metriklerini hesapla ve raporla
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred, output_dict=True)
    sensitivity = conf_matrix[1,1] / (conf_matrix[1,0] + conf_matrix[1,1])
    specificity = conf_matrix[0,0] / (conf_matrix[0,0] + conf_matrix[0,1])
    accuracy = (conf_matrix[0,0] + conf_matrix[1,1]) / len(y_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    
    print(f"---- {model_name} Sonuçları ----")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("\nClassification Report:")
    print(class_report)
    print(f"Sensitivity (Recall): {sensitivity:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"ROC AUC: {roc_auc:.2f}")
    
    # ROC eğrisini çiz
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['0', '1'],
            yticklabels=['0', '1'])
    plt.xlabel('Tamin edilen')
    plt.ylabel('Gerçek')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
evaluate_model(mlp, X_test, y_test, "MLP")
```

```python
evaluate_model(svm, X_test, y_test, "SVM")
```
